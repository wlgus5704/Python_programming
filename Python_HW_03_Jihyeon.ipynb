## HW-2: Convert words (25pt)

Write a function `play_word` that takes in a single word, then converts the word and returns it: 

- it takes the first letter of a word, 
- puts it at the end, 
- and appends `"ay"`. 

The only exception is if the first letter is a vowel `['a','e','i','o','u']`, in which case we keep the first letter as it is and just append `"hay"` to the end. 

Example: `"boot"` -> `"ootbay"`, and `"image"` -> `"imagehay"`.

It will be useful to define a list at the top of your code called `vowels`. This way, you can check if a letter `x` is a vowel with the expression `x in vowels`. 

**Example of expected result**

image.png

````python
vowels = ['a','e','i','o','u']
lst=[]
# your code here:
def play_word (word):
  lst = list(word)
  for i in range(0,len(vowels)):
    if lst[0] == vowels[i]:
      word = word+'hay'
      return word
  if lst[0] != vowels[i]:
      lst.append(lst[0])
      del lst[0]
      word = "".join(lst) + 'ay'
      return word  

````

````python
# test your code:
word1 = 'boot'
word2 = 'image'

print(f'Word1: {word1} -> New word1: {play_word(word1)}')
print(f'Word1: {word2} -> New word1: {play_word(word2)}')
````
